## GENERATE SAMPLES : 3+5 MARKS

```python
def generate_x(omega, t, x0=np.float64(0)):
    '''
    Inputs:
        omega: scalar float, the angular velocity in radians/s
        t: scalar float, time in s
        x0: scalar float, initial angle at t=0 in radians, default x0=0
    Outputs:
        x: scalar float, angle at t=t in radians, between [0, 2 pi)
    '''

    ### WRITE YOUR CODE HERE - 3 MARKS
    
    # Get angle that should be swept
    angle_to_sweep = x0 + omega*t
    
    # Get the angle b/w limits
    normalized_angle = angle_to_sweep%(2*np.pi)
    # Return answer
#     if(np.allclose(normalized_angle, 2*np.pi, rtol=1e-5)):
#         x = 0
#     else:
    x = normalized_angle
#     print(x)

    return x
```

```python
def sample_x(omega, Fs, tfinal):
    '''
    Inputs:
        omega: scalar float, the angular velocity in radians/s
        Fs: (float) sampling rate in Hz
        tfinal: (float) final time in s
    Output:
        xs: (np array, 1D float) samples of x from t=0 to t=tfinal
        ts: (np array, 1D float) samples of t from t=0 to t=tfinal, sampled uniformly
    '''

    ### WRITE YOUR CODE HERE - 5 MARKS
    
    # Sample uniformly ts
    step = 1/Fs
    ts = np.arange(0, tfinal, step, dtype=np.float32)
    
    # Get xs for given ts
    xs = np.float32(generate_x(omega, ts))
    
    return xs, ts
```

```python
def test_sample_x():
    xs,ts = sample_x(2*np.pi, 2, 2)
    assert np.all( np.isclose(xs, np.array([0., np.pi, 0., np.pi]), atol=0.001) )
    assert np.all(np.isclose(ts, np.array([0. , 0.5, 1. , 1.5]), atol=0.001) )
    print('Test passed', '\U0001F44D')

if __name__=="__main__":
    test_sample_x()
```

# DFT OF TIME SERIES : 5 MARKS

```python
def dft_xs(xs, ts, Nfft):
    '''
    Inputs:
        xs, ts: defined above. Note that ts are uniformly sampled
        Nfft: (float) window size for DFT. It may be smaller or larger than len(xs)
    Outputs:
        Xs: (np array of shape (Nfft,)) magnitude dft of xs
        fs: (np array of shape (Nfft,)) frequency value (in Hz) of corresponding index of Xs
    '''

    ### WRITE YOUR CODE HERE - 5 MARKS
    
    # Zero pad for longer window size
    if len(xs)<Nfft:
        xs_ = np.zeros((Nfft, ))
        xs_[:len(xs)] = xs
        xs = xs_
    else:
        xs = xs[:Nfft]
        ts = ts[:Nfft]
    
    # Define sampling rate
    Fs = 1/abs(ts[0]-ts[1])
    
    # Nyquist frequency
    Fnq = Fs/2
    
    # Define time points
    pnts = Nfft
    
    
    fourTime = np.array(range(0, pnts))/pnts
    fourCoefs = np.zeros((pnts, ), dtype=np.complex)
    
    for fi in range(0,pnts):
    
        # create complex sine wave
        csw = np.exp( -1j*2*np.pi*fi*fourTime )

        # compute dot product between sine wave and signal
        # Fourier coefficients
        fourCoefs[fi] = np.sum( np.multiply(xs,csw) ) / pnts
     
    # extract amplitudes
#     print(fourCoefs)
    Xs = 2*np.abs(fourCoefs)

    # compute frequencies vector
    fs = np.linspace(0,pnts,num=pnts)*Fs/pnts

    return Xs, fs
```

```python
def test_dft_xs():
    xs, ts = sample_x(20*np.pi, 10, 5)
    Nfft = 5
    Xs, fs = dft_xs(xs, ts, Nfft=Nfft)
    assert Xs.shape==(Nfft,) or Xs.shape==(Nfft,1)
    assert fs.shape==(Nfft,) or fs.shape==(Nfft,1)
    print('Test passed', '\U0001F44D')
if __name__=="__main__":
    test_dft_xs()
```

```python
def plot_Xs_(Xs,fs, freqs):
    plt.plot(fs,Xs, 'o-')
    plt.axvline(x = np.median(fs))
    plt.xlabel('frequencies')
    plt.ylabel('magnitude')
    plt.title('plot at frequencies__:'+str(freqs))
    plt.xlim((0,50))
    plt.show()
    
def changingFs():

    ### WRITE YOUR CODE HERE - 5 MARKS
    
    for freqs in np.arange(5, 35, 2.5):
        xs, ts = sample_x(20*np.pi, freqs, 5)
        xs -= np.mean(xs)
        
        Nfft = len(xs)
        Xs, fs = dft_xs(xs, ts, Nfft=Nfft)
        plot_Xs_(Xs,fs, freqs)

    return

if __name__=="__main__":
    changingFs()
```
